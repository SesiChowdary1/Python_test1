#> Plot a graph showing the relationship between PPN and number of Bedrooms.

```python
import pandas as pd
```

```python
file_path = "/home/nineleaps/data/cleaned_cap_rooms.csv"
```

```python
df = pd.read_csv(file_path)
df
```

```python
# Remove any leading or trailing spaces from column names
df.columns = df.columns.str.strip()

# Check the cleaned column names
print(df.columns)
```

```python
# Check for missing values in 'PPN' and 'Bedrooms'
print(df[df['PPN'].isnull()])
print(df[df['Beds'].isnull()])
```

```python
# Drop rows where 'PPN' or 'Bedrooms' are missing
df = df.dropna(subset=['PPN', 'Beds'])

# Confirm no missing values
print(df.isnull().sum())
```

file_path = "/home/nineleaps/data/cleaned_cap_rooms.csv"

```python
import matplotlib.pyplot as plt
```

```python


# Plot the relationship between 'PPN' and 'Beds'
plt.figure(figsize=(10, 6))
plt.bar(df['Beds'], df['PPN'], alpha=0.5)
plt.title('Relationship between PPN (Price Per Night) and Number of Beds')
plt.xlabel('Number of Beds')
plt.ylabel('Price Per Night (PPN)')
plt.grid(True)
plt.show()
```
