# Find average PPN for each city

```python
import pandas as pd
```

```python
file_path = "/home/nineleaps/Downloads/cap_rooms.csv"
```

```python
df = pd.read_csv(file_path)
```

```python
df
```

```python
# Convert PPN column to numeric, handling different formats
df['PPN'] = df['PPN'].astype(str).str.replace('₹', '', regex=False)
df['PPN'] = df['PPN'].str.replace(',', '', regex=False)
df['PPN'] = df['PPN'].str.lower().replace({'one thousand five hundred': '1500',
                                             'two thousand': '2000',
                                             'eight hundred': '800'})
df['PPN'] = pd.to_numeric(df['PPN'], errors='coerce')
```

```python
df
```

```python

# Standardize city names
# Standardize city names
df['City'] = df['City'].replace({'Hydrabad': 'Hyderabad', 'Bangalor': 'Bangalore', 'Delhi ': 'Delhi'})

df
```

```python
# Export the cleaned dataset to a new CSV file
output_file_path = "cleaned_cap_rooms.csv"  # Change the filename if needed
df.to_csv(output_file_path, index=False)

print(f"File successfully saved as {output_file_path}")
```

```python
# Group by city and calculate the average PPN
avg_ppn_per_city = df.groupby('City')['PPN'].mean().reset_index()
```
